# Transaction Anomaly Detection Prototype using Isolation Forest

Given the huge volume of transactions banks receive every day, it is difficult to detect and identify suspicious activities purely based on rule-based methods. Generation of multiple false positives or false negatives from these methods put alot of stress manpower resources. However, the mass amount of data means that we are able to pump it into a machine learning model and predict the anomaly score from the records we have. 

In transaction monitoring, we are mainly concerned with outlier detections. The goal is to separate a core of regular observations from some polluting ones.

One efficient way of performing outlier detecting in high-dimensional datasets is to use random forests. The Isolation Forest algorithm 'isolates' observations by randomly selecting a feature and then randomly selecting a split value between the max and min values of the selected feature. Since recursive partitioning can be represented by a tree structure, the number of splittings required to isolate a sample is equivalent to the path length from the root node to the terminating node.

This path length, averaged over a forest of such random trees, is a measure of normality and our decision function

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import IsolationForest
import pandas
import numpy as np
import matplotlib.pyplot as plt
import re
import os
import fileinput
import seaborn as sns

%matplotlib inline
```

The vey first step is to import our raw data into a DataFrame object for data processing. As machine learning models only take in numeric values, we need to drop columns that are not of much use and drop rows that have error string data

```python
X_train = pandas.read_excel("data.xlsx", parse_cols="A:I")
```

```python
X_train["Counterparty Country"] = [int(i[1:]) for i in X_train["Counterparty Country"]]
X_train["Counterparty"] = [int(i) for i in X_train["Counterparty"]]
```

```python
X_train.drop(['Transaction ID', 'Date of Transaction', 'Account ID', 'Account Created Date'], axis=1, inplace=True)
```

For all machine learning models, we will need two sets of data. The first set will be training data, used to train the model. As we are doing anomaly detection, the training data should be mainly clean. The second set will be test data, used to test the model accuracy. Here we have 123000+ data, with the first 120000 allocated for training, and the latter 3000 allocated for testing

```python
X_train = X_train[X_train["Transaction Amount"].apply(lambda x: type(x) in [int, np.int64, float, np.float64])]
X_train = np.nan_to_num(X_train)
X_test = X_train[120000:]
X_train = X_train[:120000]
```

```python
print np.where(np.isnan(X_train))
```

```python
clf = IsolationForest()
clf.fit(X_train)
```

As our data is multivariate (5 variables), it is impossible to plot a 5d graph for visualization. Therefore, we use a matrix scatterplot to show each pair of variables plotted against each other. 

```python
pandas.tools.plotting.scatter_matrix(pandas.DataFrame(X_train, columns=['Counterparty Country', 'Counterparty', 'Account Risk Level', 'Age of Account', 'Transaction Amount']), figsize=(8, 8), diagonal="kde")
plt.tight_layout()
plt.show()
```

To further investigate a single scatterplot, plot the scatterplot in more detail. Here we want to take a look in the relationship between transaction amount and account risk level. We can see that the density of the plot decreases drastically when the transaction amount is high.

```python
df_X_train = pandas.DataFrame(X_train, columns=['Counterparty Country', 'Counterparty', 'Account Risk Level', 'Age of Account', 'Transaction Amount'])
sns.plt.clf()
sns.lmplot("Account Risk Level", "Transaction Amount", df_X_train, hue="Counterparty Country", fit_reg=False)
sns.plt.show()
```

The decision function returns the average anomaly score of testing data. The anomaly score of an input sample is computed as the mean anomaly score of the trees in the forest. The lower the score, the more abnormal is the record.



```python
dec = clf.decision_function(X_test)
plt.ylabel('Anomaly Score')
plt.xlabel('Test Data Index')
plt.scatter(list(xrange(len(dec))), dec, s=5)
```

```python
prediction = clf.predict(X_test)
anomaly_data = [X_test[i] for i in xrange(len(X_test)) if prediction[i]==-1]
```

```python
print str(len(anomaly_data)) + " Anomalous records found out of a total " + str(len(X_test)) + " records"
```
